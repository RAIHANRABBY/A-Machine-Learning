```python
df['dep_hour']=df['Dep_Time'].apply(lambda x:int(x.split(':')[0]))

df['dep_min']=df['Dep_Time'].apply(lambda x:int(x.split(':')[1]))
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df['arrival_hour']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0]).apply(lambda x:int( x.split(':')[0]))
df['arrival_min']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0]).apply(lambda x:int( x.split(':')[1]))
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
d_list=df['Duration'].apply( lambda x:((x.replace('h','').strip('m'))).split(' '))
```

```python
for i in d_list:
    if len(i)==1:
        i.append('0')
df['Duration']=d_list.apply(lambda x : (int(x[0])*60)+int(x[1]))
```
