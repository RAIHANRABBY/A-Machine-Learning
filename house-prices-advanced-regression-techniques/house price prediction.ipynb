### divide the numbarical and non numbarical data

```python
num=[]
obj=[]
for col in x.columns:
    if x[col].dtype == np.number:
        num.append(col)
    else:
        obj.append(col)
print('number:\n',num ,end='\n\n\n')
print("object:\n",obj)
```

### remove the null value of the numberical features

```python
x[num].isnull().sum()
```

```python
x[num].describe()
```

```python
x[num]=x[num].fillna(x[num].median())
```

```python
x[num].isnull().sum()
```

```python
x[num].describe()
```
